### **Pregunta 2 -** ¿Hubo un cambio significativo en la distribución del PIB por región comparando los valores de 2002 y 2020?

#### Tratando los datos para generar la visualización

```python
df_pib_2002 = df_pib.query('año == 2002')[['region','pib']]
df_pib_2020 = df_pib.query('año == 2020')[['region','pib']]

df_pib_2002 = df_pib_2002.groupby('region').sum().sort_values('pib',ascending=False)
df_pib_2020 = df_pib_2020.groupby('region').sum().sort_values('pib',ascending=False)

display(df_pib_2002,df_pib_2020)
```

#### Generando el gráfico

```python
fig, axs = plt.subplots(1,2,figsize=(14,6))

# Define the color list for the 5 regions
colores =[AZUL3, NARANJA1, AZUL5, VERDE1, GRIS5]

plt.suptitle('PRODUCTO INTERNO BRUTO (PIB) POR REGION', fontsize=18,
             color=GRIS1, ha='right', y=1.05)

# Grafico 1 (for 2002)

ptc,text,_ = axs[0].pie(df_pib_2002['pib'],
                         labels=df_pib_2002.index,
                         autopct='%.1f%%',pctdistance=0.6,textprops={'size':12,
                                                                     'fontweight':'bold','color':BLANCO},
                        colors=colores)
axs[0].annotate('En 2002',xy=(50,350),size=12,xycoords='axes points',
                bbox=dict(boxstyle='round',fc= BLANCO,ec=GRIS3))
for i, p in enumerate(ptc):
  text [i].set_color(p.get_facecolor())


  # Grafico 2 (for 2020)

ptc,text,_ = axs[1].pie(df_pib_2020['pib'],
                         labels=df_pib_2020.index,
                         autopct='%.1f%%',pctdistance=0.6,textprops={'size':12,
                                                                     'fontweight':'bold','color':BLANCO},
                        colors=colores)
axs[1].annotate('En 2020',xy=(50,350),size=12,xycoords='axes points',
                bbox=dict(boxstyle='round',fc= BLANCO,ec=GRIS3))
for i, p in enumerate(ptc):
  text [i].set_color(p.get_facecolor())

plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(14,6))

# Define the color list for the 5 regions
colores =[AZUL3, NARANJA1, AZUL5, VERDE1, GRIS5]

plt.suptitle('PRODUCTO INTERNO BRUTO (PIB) POR REGION', fontsize=18,
             color=GRIS1, ha='right', y=1.05)

# Grafico 1 (for 2002)

ptc,text,_ = axs[0].pie(df_pib_2002['pib'],
                         labels=df_pib_2002.index,
                         autopct='%.1f%%',pctdistance=0.6,textprops={'size':12,
                                                                     'fontweight':'bold','color':BLANCO},
                        colors=colores)
axs[0].annotate('En 2002',xy=(50,350),size=12,xycoords='axes points',
                bbox=dict(boxstyle='round',fc= BLANCO,ec=GRIS3))
for i, p in enumerate(ptc):
  text [i].set_color(p.get_facecolor())


  # Grafico 2 (for 2020)

ptc,text,_ = axs[1].pie(df_pib_2020['pib'],
                         labels=df_pib_2020.index,
                         autopct='%.1f%%',pctdistance=0.6,textprops={'size':12,
                                                                     'fontweight':'bold','color':BLANCO},
                        colors=colores)
axs[1].annotate('En 2020',xy=(50,350),size=12,xycoords='axes points',
                bbox=dict(boxstyle='round',fc= BLANCO,ec=GRIS3))
for i, p in enumerate(ptc):
  text [i].set_color(p.get_facecolor())
   # anotacion de coclucion

axs[0].annotate('la region sudeste fue la unica que redujo\n su participacion en el PIB.',
                   xy=(180,245),xycoords='axes points',xytext=(280,295),
                   textcoords='axes points',
                   bbox=dict(boxstyle='round',fc=BLANCO,ec=GRIS3),
                   size=10,arrowprops=dict(arrowstyle='->',fc=GRIS1,
                                           connectionstyle='arc,angleA=-90,angleB=0,armA=30,armB=30,rad=5'))

axs[1].annotate('n\n',
                   xy=(140,245),xycoords='axes points',xytext=(-50,295),
                   textcoords='axes points',
                   size=10,arrowprops=dict(arrowstyle='->',fc=GRIS1,
                                           connectionstyle='arc,angleA=-90,angleB=0,armA=30,armB=-30,rad=5'))
plt.show()
```

### **Pregunta 3 -** ¿Cuál es la participación del estado de Minas Gerais en el PIB de todo Brasil en el año de 2020?

#### Tratando los datos para generar la visualización

### **Pregunta 4 -** ¿Cómo está distribuido el PIB del estado de Bahía en 2020 separado por impuestos líquidos y los valores agregados brutos de los bienes y servicios producidos?

#### Tratando los datos para generar la visualización

```python
bahia= df_pib.query('estado == "BA" & año == 2020')[['pib','impuesto_liquido','va']]
bahia.rename(columns={'pib':'PIB','impuesto_liquido':'Impuestos_Liquidos','va':'Valor_Agregado_bruto'}, inplace=True)
bahia
```

```python
bahia = df_pib.query('estado == "BA" & año == 2020')[['pib','impuesto_liquido','va']]
bahia.rename(columns={'pib':'PIB','impuesto_liquido':'Impuestos_Liquidos','va':'Valor_Agregado_bruto'}, inplace=True)
bahia = bahia.melt(var_name='Indicadores',value_name='Valores')
bahia
```

```python
bahia.loc[bahia['Indicadores']=='Impuestos Liquidos','Valores']=bahia['Valores']*-1
bahia['Valores_str']=(bahia['Valores']/1e9).map('R${:,.2f}Bi'.format)
bahia['Medidas']=['absolute','relative','total']
bahia
```

#### Generando el gráfico

```python
import plotly.graph_objects as go
fig = go.Figure(go.Waterfall(name='',orientation='v',
                measure=bahia['Medidas'],
                             x=bahia['Indicadores'],y=bahia['Valores'],
                             text=bahia['Valores_str'],
                             textposition='outside',
                             connector_line=dict(color=GRIS3),
                             totals=dict(marker=dict(color=AZUL3)), # Corrected from total_marker
                             decreasing_marker=dict(color=ROJO1)))
fig.update_layout(width=1000,height=500,font_family='DejaVu Sans',
                  font_size=15, font_color=GRIS2,title_font_color=GRIS1,
                  title_font_size=24,
                  title_text='Distribucion del producto interno bruto ',
                  title_pad_l=50,plot_bgcolor=BLANCO, hovermode='closest')

fig.update_yaxes(showticklabels=False)
fig.update_traces(hovertemplate='<b>%{x</b>=%{text}')


fig.show()
```

<font color=green size=4> Conclusiones: </font>

---

Un gráfico de cascada es ampliamente utilizado en datos económicos y financieros debido a su naturaleza de representar de manera simple y directa la acumulación y/o sustracción del total en un período determinado o en diferentes categorías.

Como sugerencia, podemos analizar estos valores desde otro estado de tu preferencia, o incluso separar los valores agregados por tipo de servicio mostrando la agregación de cada uno.

**Observación:** Para ayudar en la comprensión del gráfico, los bienes y servicios finales que componen el PIB se miden al precio al que llegan al consumidor. De esta manera, también se tienen en cuenta los impuestos sobre los productos comercializados.

## **2.2 - Añadiendo más incrementos/disminuciones**

---



### **Pregunta 5 -** ¿Cuál es la evolución anual del PIB del estado de Río de Janeiro entre los años de 2010 a 2020?

#### Tratando los datos para generar la visualización

```python
anos =[x for x in range(2010,2021)]
rio = df_pib.query('estado=="RJ" & año in @anos')[['año','pib']]
#rio.rename(columns={'pib':'PIB'},inplace=True
rio = rio.reset_index(drop=True)
rio
```

```python
rio['variacion_pib'] = rio['pib'].diff().fillna(rio['pib'].astype('int64'))

rio ['medidas'] =['absolute'] +['relative']*10
rio
```

```python
final_pib_value = rio['pib'].iloc[-1] # Get the last PIB value

actualizacion = pd.Series({'año':'Total',
                           'pib':final_pib_value,
                           'variacion_pib':final_pib_value,
                           'medidas':'total'})

rio = pd.concat([rio, actualizacion.to_frame().T], axis=0, ignore_index=True)
rio = rio.drop(columns=['pib'])
rio
```

```python
#final_pib_value = rio['pib'].iloc[-1] # Get the last PIB value

#actualizacion = pd.Series({'año':'Total',
#                           'pib':final_pib_value,
#                           'variacion_pib':final_pib_value,
#                           'medidas':'total'})

#rio = pd.concat([rio, actualizacion.to_frame().T], axis=0, ignore_index=True)
#rio = rio.drop(columns=['pib'])
rio ['variacion_pib_texto']= (rio['variacion_pib']/1e9).map('R${:,.2f}Bi'.format)
rio.rename(columns={'año':'Año',
                    'variacion_pib':'Variacion del PIB',
                    'variacion_pib_texto':'Variacion del PIB (texto)' }, inplace=True)
rio
```

#### Generando el gráfico

```python
import plotly.graph_objects as go
import numpy as np # Import numpy

fig = go.Figure(go.Waterfall(
    name='',orientation='v',
    measure=rio['medidas'],
    x=rio['Año'], # Uncommented and corrected x argument
    y=rio['Variacion del PIB'],
    text=rio['Variacion del PIB (texto)'], # Corrected text argument
    textposition='outside',
    connector_line=dict(color=GRIS3),
    totals=dict(marker=dict(color=AZUL3)), # Corrected property name
    decreasing_marker=dict(color=ROJO1)))

fig.update_layout(width=1300,height=600,
                   font=dict(family='DejaVu Sans', size=15, color=GRIS2), # Corrected font properties
                   title_font_color=GRIS1,
                   title_font_size=24,
                   title_text='Variacion del PIB del Estado de Río de Janeiro (2010-2020)', # Updated title
                   plot_bgcolor=BLANCO,yaxis_range=[0,850e9])
fig.update_yaxes(showticklabels=False)
fig.update_xaxes (tickmode='array',tickvals=np.arange(0,len(rio['Año'])),ticktext=rio['Año']) # Corrected np.arrange to np.arange and adjusted tickvals length

fig.show()
```

Vamos a trabajar con una función para la escritura de texto utilizando colores diferentes.

> (Adaptada de https://github.com/empathy87/storytelling-with-data)



```python
# Función para escribir texto usando diferentes colores. No existe una fución
# nativa para hacerlo. (Adaptado de https://github.com/empathy87/storytelling-with-data)
from matplotlib import transforms

def texto_a_color(x, y, texto, colores, esp=20, ax=None, **kw):
    colores = list(reversed(colores))
    t = ax.transData
    canvas = ax.figure.canvas

    for i, linha in enumerate(reversed(texto.split('\n'))):
        frases = linha.split('||')
        for s, cor in zip(frases, colores[i]):
            texto = ax.text(x, y, s, color=cor, transform=t, **kw)
            texto.draw(canvas.get_renderer())
            ex = texto.get_window_extent()
            t = transforms.offset_copy(texto._transform, x=ex.width,
                                       units='dots')

        t = transforms.offset_copy(ax.transData, x=0, y=(i + 1) * esp, units='dots')
```

#### Generando el gráfico

```python
ax,colores =grafico_va()

ax.legend().remove()
texto_a_color(x=0,y=4.4,
             texto='$\\bf{Servicios}$ || | || $\\bf{Industria}$ || | || $\\bf{ADESPSS}$ || | || $\\bf{Agropecuario}$',
              colores=[[colores[0],GRIS3,colores[1],GRIS3,colores[2],GRIS3,colores[3]]],
              ax=ax,fontsize=12)

texto_a_color(1.)
fig.show()
```
