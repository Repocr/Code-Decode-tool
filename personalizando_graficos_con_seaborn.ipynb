```python
def generar_grafico(palette):
    fig,ax = plt.subplots(figsize=(8,4))
    ax =sns.barplot(data=top_10,x='Total',y=top_10.Pais,orient='h',palette=palette,legend=False)
    ax.set_title('Top 10 paises con mayor migracion a Canada\n 1980-2013',loc='left',fontsize=16)
    ax.set_xlabel('Numero de inmigrantes',fontsize=14)
    ax.set_ylabel('')

    plt.show()
```

```python
generar_grafico('Blues_r')
```

```python
generar_grafico('rocket')
```

```python
generar_grafico('coolwarm')
```

```python
generar_grafico('Paired')
```

```python
generar_grafico('tab10')
```

```python
def generar_grafico2(palette):
    fig,ax = plt.subplots(figsize=(8,4))
    ax.set_frame_on(False)
    ax.xaxis.set_visible(False)
    ax.tick_params(axis='y',which='both',size=0)
    ax =sns.barplot(data=top_10,x='Total',y=top_10.Pais,hue=top_10.Pais,orient='h',palette=palette,legend=False)
    ax.set_title('Top 10 paises con mayor migracion a Canada\n 1980-2013',loc='left',fontsize=16)
    ax.set_xlabel('Numero de inmigrantes',fontsize=14)
    ax.set_ylabel('')
    for i,j in enumerate(top_10['Total']):
        ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
    plt.show()
```

```python
generar_grafico2('rocket')
```

```python
generar_grafico2('coolwarm')
```

```python
generar_grafico2('Paired')
```

```python
generar_grafico2('tab10')
```

```python
def generar_grafico3(palette):
    fig,ax = plt.subplots(figsize=(8,4))
    ax.set(xticklabels=[])
    ax =sns.barplot(data=top_10,x='Total',y=top_10.Pais,hue=top_10.Pais,orient='h',palette=palette,legend=False)
    ax.set_title('Top 10 paises con mayor migracion a Canada\n 1980-2013',loc='left',fontsize=16)
    ax.set_xlabel('Numero de inmigrantes',fontsize=14)
    ax.set_ylabel('')
    sns.despine()
    for i,j in enumerate(top_10['Total']):
        ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')
    plt.show()
```

```python
generar_grafico3('tab10')
```

```python
sns.set_theme(style='darkgrid')
generar_grafico3('tab10')
```

```python
sns.set_theme(style='whitegrid')
generar_grafico3('tab10')
```

```python
sns.set_theme(style='white')
generar_grafico3('tab10')
```

```python
sns.set_theme()
sns.set_palette('Dark2')
df_indexed = df.set_index('Pais')
años = [str(year) for year in range(1980, 2014)]
fig, ax = plt.subplots(figsize=(10, 5))

# Filter for South American countries and year columns
south_america_countries = ['Brasil', 'Argentina', 'Perú', 'Colombia']
df_south_america = df_indexed.loc[south_america_countries, años]

# Melt the DataFrame to long format
df_south_america_melted = df_south_america.reset_index().melt(id_vars='Pais', var_name='Año', value_name='Numero de Inmigrantes')

# Convert 'Año' back to numeric for plotting
df_south_america_melted['Año'] = pd.to_numeric(df_south_america_melted['Año'])

# Plot using the melted DataFrame
ax = sns.lineplot(data=df_south_america_melted, x='Año', y='Numero de Inmigrantes', hue='Pais', lw=3)


ax.set_title('Inmigración de los principales países de América del Sur\nal Canadá de 1980 a 2013', loc='left', fontsize=20)
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.legend(title='Países', loc='upper right', bbox_to_anchor=(1.18, 1.02))

plt.show()
```
